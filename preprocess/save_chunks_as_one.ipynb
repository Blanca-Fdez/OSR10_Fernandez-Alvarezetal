```python
# Description: This script merges the chunks of harmonic results and residuals into a single file.
import xarray as xr
import os
import sys

sys.path.append('/home/bfernandez/Escritorio/MHW/Estadistica')
import blanca_tools         as btools

# Define output folder
out_fold = "/home/bfernandez/Escritorio/OSR10/harmonics_trends/"
fold = "/home/bfernandez/Escritorio/OSR10/data/"

harmonic_files = btools.dirtodict(out_fold + "/results")[".files"]


```

```python
if harmonic_files:
    ds_harmonics = xr.open_mfdataset(harmonic_files, combine="by_coords")
    ds_harmonics.to_netcdf(os.path.join(out_fold, "harmonic_results_combined.nc"))
    print("Combined harmonic results saved.")
```

```python
residual_files = btools.dirtodict(out_fold + "/residuals")[".files"]
ds_residual = xr.open_mfdataset(residual_files, combine="by_coords")
```

```python
ds_residual.to_netcdf(f"{out_fold}/residuals_combined.nc")
```

```python
if residual_files:
    ds_residuals = xr.open_mfdataset(residual_files, combine="by_coords")
    ds_residuals.to_netcdf(os.path.join(out_fold, "residuals_combined.nc"))
    print("Combined residuals saved.")
else:
    print("No residuals found.")
```
